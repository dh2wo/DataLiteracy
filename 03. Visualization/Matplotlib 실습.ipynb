# 선도표(line plot) 그리기

## 기본적인 그래프를 그리기

<pre>
plt.plot([x], y, [fmt], [x2], y2, [fmt2] ..., **kwargs)
fmt(formatting): 색상, 마커, 선 스타일과 같은 기본 서식 설정
(하나의 plot에 여러 개의 좌표와 fmt를 설정하면 복수개의 그래프를 그릴 수 있다.)
</pre>
- plot() 함수의 인자료 표시할 데이터를 전달한 후 **show()** 함수를 통해 출력한다.
- plot() 함수에 전달되는 리스트는 자동으로 y축의 값으로 설정된다.
- x 축의 값은 0으로 시작해서 1의 간격으로 자동으로 설정된다.

```python
plt.plot([0, 1, 2, 3], 'ro')
plt.show()
```

- 두개의 리스트를 전달함으로써 각각 x축의 값, y축의 값으로 설정된 그래프를 그릴 수도 있다.

```python
plt.plot([2, 4, 6, 8], [1, 16, 4, 9]) # x축: [2, 4, 6, 8], y축: [1, 16 ,4, 9]
plt.show()
```

```python
plt.plot([2, 4, 6, 8], [1, 16, 4, 9], 'ro') # x축: [2, 4, 6, 8], y축: [1, 16 ,4, 9]
plt.axis([0, 10, 0, 20]) # 축 범위 [xmin, xmax, ymin, ymax]
plt.show()
```

## 스타일
- plot() 함수에 다양한 키워드 인수를 전달하여 스타일을 바꿀 수 있다.
    - linestyle: solid, dashed, dotted, dash-dot

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2,
        color='red',
        linestyle='dashed',
        linewidth=2.5,
        marker='o',
        markeredgecolor='green',
        markerfacecolor='yellow',
        markersize=9.3
        )
plt.show()
```

- 상징(각 스타일을 나타내는 단축 문자열)을 이용해서 문자열 형태로 설정할 수도 있다.

```python
x = np.linspace(0, np.pi*2, 10)

plt.plot(x, np.sin(x), 'ro--', x, np.cos(x), 'g^-.', x, -np.sin(x), 'bx:') # color, makers, line style
plt.xlim(0, np.pi*2) # xlim(): x축의 범위만 지정
plt.show()
```

- 주요 요소들에 대한 상징은 다음 표와 같다.


```python
plt.plot([1, 2, 3, 4], [5, 20, 24, 56], 'bx--')
plt.title('Line Plot')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.show()
```

```python
data = np.arange(0, 5, 0.2)
plt.plot(data, data, 'r--', data, data**2, 'bs', data, data**3, 'g^:')
plt.show()
```

- 기본적인 스타일 외에도 인수 이름을 지정하여 다양한 스타일 지정이 가능하다.

# 서브플롯 그리기

subplot과 subplots 모두 한번에 여러 그래프를 보여주기 위해 사용되는 코드<br>
subplot는 일일이 다 설정해줘야하지만
subplots는 한번에 설정해 줄 수 있다.

- plt.subplot() 함수를 이용해서 하나의 figure에 여러 개의 그래프를 그릴 수 있다.
- 주요 인자
    - nrows, ncols, index를 지정하는 3가지 정수를 이용하여 서브플롯의 위치를 설정한다.

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2*np.pi * x2)

# plt.figure(figsize=(11, 4))

plt.subplot(2, 1, 1) # nrows=2, ncols=1, index=1: 2행 1열 첫번째 그림
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time(s)')
plt.ylabel('Damped')

plt.subplot(2, 1, 2) # nrows=2, ncols=1, index=1: 2행 1열 두번째 그림
plt.plot(x2, y2, '.-')
plt.title('2st Graph')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

plt.tight_layout() # 서브플롯의 레이아웃을 자동으로 맞춰준다.
plt.show()
```

- plt.subplots() 함수를 이용해서 서브플롯을 그릴 수 있다.
- subplots() 함수는 Figure와 Axes 객체를 반환한다.
    - Axes 클래스: 공간이 확보된 이미지 영역 정보를 가지고 있다.
- 주요 인자
    - nrows, ncols: 정수 값으로 서브플롯의 행, 열수 지정
    - sharex, sharey: boolean 값으로(defalut: False) x, y축 사이의 공유 설정

```python
fig, ax = plt.subplots(1, 4, sharey=True) # 1행 4열

fig.set_size_inches((16, 4)) # figure 사이즈 설정

ax[0].scatter(np.random.rand(100), np.random.rand(100), c='cyan', edgecolor='red', alpha=0.5) # scatter: 산점도
ax[1].scatter(np.random.rand(100), np.random.rand(100), c='magenta', edgecolor='green', alpha=0.5)
ax[2].scatter(np.random.rand(100), np.random.rand(100), c='yellow', edgecolor='blue', alpha=0.5)
ax[3].scatter(np.random.rand(100), np.random.rand(100), c='white', edgecolor='black', alpha=0.5)

ax[0].set_title('subplot 0')
ax[1].set_title('subplot 1')
ax[2].set_title('subplot 2')
ax[3].set_title('subplot 3')
```

```python
fig, ax = plt.subplots(2, 2)

x = np.arange(0, 10, 0.01)
ax[0][0].plot(x, np.sin(x))
ax[0][1].plot(x, np.cos(x))
ax[1][0].plot(x, np.tanh(x))
ax[1][1].plot(x, np.sin(x)**2)

plt.show()
```

# 막대 그래프(bar chart) 그리기
- plt.bar() 함수를 이용해서 막대 그래프를 그릴 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,20)
y = x*2

plt.bar(x, y, label='Blue Bar', color='blue' , edgecolor='yellow', linewidth=3)
plt.title('Bar Chart Example')
plt.xlabel('bar number')
plt.ylabel('bar height')
x_ticks = (np.arange(len(x))) # x의 길이(20)만큼 눈금 설정
plt.xticks(x_ticks, x_ticks+1) # (축 인덱스, 새로운 축 번호 값)
plt.legend() # 범례
plt.show()
```

- plt.barh() 함수를 이용해서 수평 막대 그래프를 그릴 수 있다.

```python
x = np.arange(0,20)
y = x*2

plt.barh(x, y, label='Blue Bar', color='blue' , edgecolor='yellow', linewidth=3)
plt.title('H-Bar Chart Example')
plt.xlabel('bar height')
plt.ylabel('bar number')
plt.yticks(np.arange(len(x))) # x의 길이(20)만큼 눈금 설정
plt.legend() # 범례
plt.show()
```

- matplotlib 색상표


```python
labels = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(men_means))
width = 0.35 # 막대 그래프 폭
plt.bar(x - width/2, men_means, width=width, color='dodgerblue', label='Men')
plt.bar(x + width/2, women_means, width=width, color='violet', label='Women')

plt.title('Score by group and gender')
plt.xlabel('City')
plt.xlabel('Score')
plt.xticks(x, labels) # 눈금
plt.legend()
plt.show()
```

# 히스토그램 그리기
- plt.hist() 함수를 이용해서 히스토그램을 그릴 수 있다.
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 속하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency table)라고 한다.
- 히스토그램은 도수 분포표에서 구간별 빈도수를 막대 그래프로 나타낸 것이다.

```python
y = np.random.randn(1000) # 평균이 0, 표준편자가 1인 표준정규분포로부터 난수 1000개 생성

x = np.arange(len(y))
plt.bar(x, y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y, bins=50) # 표현할 데이터를 50개의 구간으로 나눈다.
plt.title('histogram')
plt.show()

plt.hist(y, bins=20, cumulative=True) # 누적 빈도수
plt.title('Cumulative Histogram')
plt.show()
```

# 원 그래프(pie chart) 그리기
- plt.pie() 함수를 이용해서 원 그래프를 그릴 수 있다.
- pie() 함수에 전달하는 값은 굳이 비율값을 전달하지 않고 일반적인 숫자를 숫자 값을 전달하더라도 값들의 총 합에 대한 비율을 알아서 계산해서 표시한다.

```python
labels = ['Python', 'Java', 'C']
# autopct: wedge 위에 표시할 레이블 형식(default: None)
plt.pie([0.3, 0.35, 0.4], labels=labels, autopct='%.1f%%') # %%: 숫자 뒤에 %
plt.title('Popular Language Ranking')
plt.show()
```

```python
sections = [0.3, 0.35, 0.4]
labels = ['Python', 'Java', 'C']
color = ['coral', 'dodgerblue', 'yellowgreen']

plt.pie(sections, # 조각의 크기를 설정하는 값
        explode=(0.1, 0, 0), # 조각을 중심으로부터 얼마만큼 떨어지게 할 것인지 설정 (첫 번째 조각이 반지름의 10%만큼 벗어남을 의미)
        labels=labels, # 조각의 이름을 표시
        colors=color, # 조각의 색
        autopct='%.2f%%', # 비율을 표시하는 형식 문자열을 지정
        shadow=True, # 그림자 표시
        startangle=90, # 원형의 시작 부분을 각도로 표시. (기본값은 0, 3시 방향)
        counterclock=False, # 반시계방향으로 순서대로 표시할지 결정(기본값은 True)
        rotatelabels=True # 조각의 이름을 회전시킨다. (기본값은 False)
       )

plt.title('Pie Chart Example 2')
plt.show()
```

# 박스 플롯 그리기
- plt.boxplot() 함수는 많은 데이터에서 데이터의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(Outlier)가 있는지 검수하기 수월하다.
    - 최소값: 제 1사분위에서 1.5IQR(Integer Quantile Range: 사분위수범위,Q3-Q1)을 뺀 위치
    - 제 1사분위(Q1): 중앙값 기준으로 하위 50%중의 중앙값. 즉, 전체 데이터중 하위 25%에 해당하는 값
    - 제 2사분위(Q2): 50%의 위치로 중앙값(median)을 의미. 데이터의 정 가운데 순위에 해당하는 값
    - 제 3사분위(Q3): 중앙값 기준으로 상위 50%중의 중앙값. 즉, 전체 데이터 중 상위 25%에 해당하는 값
    - 최대값: 제 3사분위에서 1.5IQR값을 더한 위치
    - IQR(사분위범위) :Q3 - Q1
    - 최대값, 최소값 범위를 벗어나는 값을 이상치로 데이터로 간주한다.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0,2.0,1000) # 평균이 0이고 표준편차가 2인 난수 1000개
data_b = np.random.normal(-3.0,1.5,500) # 평균: -3,표준편차:1.5 난수 500개
data_c = np.random.normal(1.2,1.5,1500)

plt.boxplot([data_a,data_b,data_c])
plt.ylim(-10,10)
plt.xlabel('Data Type')
plt.ylabel('Value')

plt.show()

# 밑의 그림은 상자수염그림이라고도 불린다.
```

```python
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
print(f'np.min():{np.min(data_a):.3f}, np.median():{np.median(data_a):.3f}, np.max():{np.max(data_a):.3f}')
quantile = np.quantile(data_a,[0,0.25,0.5,0.75,1]) # quantile:분위수 # 0%, 25%, 50%, 75%, 100%
print(quantile)
IQR =quantile[3]-quantile[1]
print('IQR: ',IQR)
print('MAX: ',quantile[3]+1.5*IQR)
print('MIN: ',quantile[1]-1.5*IQR)
```

```python
np.random.seed(10)
spread = np.random.rand(50) * 100 #[0,100) 사이의 값 50개
center= np.ones(25) * 50 # 50(중앙값) 25개
flier_high = np.random.rand(10) * 100 + 100 #[100,200) 사이의 값 10개
flier_low = np.random.rand(10) * 100 - 100 #[100,0) 사이의 값 10개

data = np.concatenate([spread, center, flier_high, flier_low])
print(data.shape)

plt.boxplot(data)
plt.title("Box Plot - Outlier Check")
plt.show()
```

## 매체별 광고 투자금액에 따른 상품 판매 효과 비교
- 선 그래프로 표현한다.
- 광고 매체는 tv와 google_ads 두가지로 한다.
- x(광고 투자금)값은 5~45사이의 5간격으로 만들어진 연속된 값(tv와 google_ads 공통 사용)
- tv의 y(상품 판매 금액)값은 1~64사이의 랜덤 정수(seed는 10으로 고정, 오름차순으로 정렬) 
- google_ads의 값은 1~74 사이의 랜덤 정수(seed는 10으로 고정,오름차순으로 정렬)

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 폰트 셋팅
matplotlib.rcParams['font.size'] = 12 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus']= False # 유니코드가 '-'기호로 설정되는 폰트가 깨지는 것을 방지
plt.style.use('default')
```

```python
np.random.seed(10)

x_invest = np.arange(5,46,5)
tv = np.sort(np.random.randint(1,65,9))
google_ads = np.sort(np.random.randint(1,75,9))

plt.plot(x_invest,tv,'bs-',label='TV')
plt.plot(x_invest,google_ads,'ro-',label='Google ads')
plt.title('compare')
plt.xlabel('admoney',labelpad=10) #labelpad가 뭐지
plt.ylabel('sales',labelpad=10)
plt.legend()
plt.show()
```

# 국가별 주류 소비량 상위 20개국 비교
- 막대 그래프로 표현한다.
- 주류 소비량은 '맥주','증류주','와인' 소비량의 합계로 산출한다.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('../02. Pandas/국가별 알콜섭취 데이터 분석/drinks.csv')
```

```python
column={'country' : '국가',
'beer_servings' : '맥주',
'spirit_servings' : '증류주',
'wine_servings' : '와인',
'total_litres_of_pure_alcohol' : '알코올',
'continent' : '대륙'}
df.rename(columns=column,inplace=True)
```

```python
df['주류'] = df.loc[:,'맥주':'와인'].sum(axis=1)
df
```

```python
top_20 = df.sort_values('주류',ascending=False).head(20)
top_20
```

```python
x = np.arange(len(top_20))
y = np.array(top_20['주류'])
x_label = top_20['국가']

plt.figure(figsize=(11,4))
plt.bar(x,y,width=0.5)
plt.title('top_20Country of drink')
plt.xlabel('Country')
plt.ylabel('alcohol consumption',labelpad=20)
plt.xticks(x,x_label,rotation=40)
plt.show
```
