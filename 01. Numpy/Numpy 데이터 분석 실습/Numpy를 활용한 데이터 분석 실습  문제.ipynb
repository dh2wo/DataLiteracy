# [문제] BMI 지수 계산 및 과체중 이상인 BMI 값 확인
- 10명에 대한 키와 몸무게가 들어있는 'height_weight.txt'를 읽어 각 사람별 BMI 지수를 구하고 비만 전단계 이상인 BMI 값 확인하기

<pre>
 BMI 지수 = 몸무게(kg) / 키(M)의 제곱
 체질량지수의 범위 값에 따른 비만도 값 지정
 18.5 미만 ▶ 저체중
 18.5 ~ 22.9 ▶ 정상
 23 ~ 24.9 ▶ 비만 전단계
 25 ~ 29 ▶ 1단계 비만
 30 ~ 34.9 ▶ 2단계 비만
 35이상 ▶ 고도비만
</pre>

```python
np.set_printoptions(suppress=True, precision=2)
```

```python
data = np.loadtxt('height_weight.txt', delimiter=',', dtype='float') # (파일이름, 분류자, 데이터타입)
print(data)

bmi = data[1] / (data[0] * 0.01) ** 2
print(bmi)

print('BMI 23이상인 사람들의 BMI')
print(bmi[bmi >= 23])
```

    - 성별: 1(남자), 2(여자)

    - 고혈압 당뇨병 진료여부: 1(고혈압 및 당뇨진료내역 있음), 2(고혈압 진료내역 있음), 3(당뇨 진료내역 있음), 4(고혈압 및 당뇨 진료내역 없음)
    - 정상혈압: 수축기 혈압 120nnHg 미만, 이완기 혈압 80mmHg 미만

* 전체 레코드 수와 상위 5개 데이터 확인
* 데이터 모양 확인
* 여성과 남성의 데이터 수 확인
* 여성과 남성의 수축기혈압과 이완기혈압의 평균값 확인
* 정상 혈압 남자 및 여자 수 확인
* 당뇨병 질환(공복혈당 수치 126mg/dL 이상)을 가지고 있는 사람들의 평균 체질량지수

```python
data = np.loadtxt("국가건강검진_혈압혈당데이터.csv", delimiter=",", dtype=float, skiprows=1)
```

```python
# 전체 레코드 수와 상위 5개 데이터 확인
print(f'전체 레코드수: {data.shape[0]: ,}')
print(f'상위 5개 데이터:\n {data[:5]}')
```

```python
# 데이터 모양 확인
print(data.shape[1])
```

```python
# 여성과 남성의 데이터 수 확인
male = data[data[:,0] == 1]
female = data[data[:,0] == 2]
# print(f'남성의 데이터 수 : {len(male)}')
# print(f'여성의 데이터 수 : {len(female)}')
print(f'남성의 데이터 수 : {sum(data[:, 0] == 1)}개')
print(f'남성의 데이터 수 : {sum(data[:, 0] == 2)}개')
```

```python
# 여성과 남성의 수축기혈압과 이완기혈압의 평균값 확인
# 수축기혈합: SBP, 이완기혈합: DBP
male_sbp_mean = np.mean(data[data[:, 0]==1][:, 2])
print(f'남성의 수축기혈압 평균: {male_sbp_mean: .2f}')
male_dbp_mean = np.mean(data[data[:, 0]==1][:, 2])
print(f'남성의 이완기혈압 평균: {male_dbp_mean: .2f}')
female_sbp_mean = np.mean(data[data[:, 0]==2][:, 2])
print(f'여성의 수축기혈압 평균: {female_sbp_mean: .2f}')
female_dbp_mean = np.mean(data[data[:, 0]==2][:, 2])
print(f'여성의 이완기혈압 평균: {female_dbp_mean: .2f}')

# female_s = np.mean(female[:, 2])
# print(f'여성의 수축기혈압 평균: {female_s}')
# female_e = np.mean(female[:,3])
# print(f'여성의 이완기혈압 평균: {female_e}')
```

```python
# 정상 혈압 남자 및 여자 수 확인
# 정상혈압: 수축기 혈압 120nnHg 미만, 이완기 혈압 80mmHg 미만
nbp_masking = np.logical_and(data[:, 2] < 120, data[:, 3] < 80)
nbp_data = data[nbp_masking]
print(f'정상혈합 남자 수: {sum(nbp_data[:, 0] == 1): ,}명')
print(f'정상혈합 여자 수: {sum(nbp_data[:, 0] == 2): ,}명')
```

```python
# 당뇨병(diabetes) 질환(공복혈당 수치 126mg/dL 이상)을 가지고 있는 사람들의 평균 체질량지수
diabetes = data[data[:, 4] > 126]
print(f'당뇨병 질환 환자의 평균 BMI: {np.mean(diabetes[:,-1]): .2f}')
```

# [문제] 시에틀 강수량 데이터 분석
- 'Seattle2014.csv' 파일 내용 읽기
- 강수량('PRCP')이 많은 날 상위 10일 확인
- 1월 강수량만 출력
- 1월에 내린 강수량의 합, 평균강수량 확인
- 1년동안 강수량이 40미만으로 내린 날이 몇일인가?

```python
data = np.loadtxt('Seattle2014.csv', delimiter=',', dtype=object, skiprows=1) # dtype=object: 전체 데이터를 문자열 형태로
```

```python
# 강수량('PRCP')이 많은 날 상위 10일 확인
prcp = data[:, 3].astype(int)
top_prcp = np.argsort(prcp)[::-1][:10]
print(top_prcp)
print(f'강수량이 많은날 상위 10일:\n {data[top_prcp]}')
```

```python
# 상위 강수량과 날짜도 같이 출력
prcp = data[:, [2, 3]].astype(int)
prcp = prcp[prcp[:, 1].argsort()][::-1][:10]
print(prcp)
```

```python
# 1월 강수량만 출력
prcp = data[:, [2,3]]
prcp_month = np.array([[date[4:6], prcp] for date, prcp_month in prcp])
jan_prcp = prcp_month[prcp_month[:, 0] == '01'][:, 1]
jan_prcp = prcp_month[prcp_month[:, 0] == '01'][:, 1].astype(int)
     
print(jan_prcp)
```

```python
# 1월에 내린 강수량의 합, 평균강수량 확인
print(f'1월 강수량 합: {np.sum(jan_prcp)}')
print(f'1월 강수량 합: {np.mean(jan_prcp):. 2f}')
```

```python
# 1년동안 강수량이 40미만으로 내린 날이 몇일인가?
prcp = data[:, 3].astype(int)
under_40 = prcp[prcp < 40]
print(f'40 미만으로 내린 날 수: {len(under_40)}일')
```

# [문제] 영화 평점 분석
- 'ratings.dat' 파일 내용 읽기 (데이터 형식: user_id::item_id::rating::timestamp)
- 상위 10개 데이터 출력하기
- 데이터 정보 확인(크기, 차원)
- 전체 데이터 평점 구하기
- 사용자 아이디 1인 데이터 출력하기
- 사용자 아이디 1인 사람의 평점 평균 구하기
- 각 사용자별 평점 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
- 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기

```python
data = np.loadtxt('ratings.dat', delimiter='::', dtype=int)
```

```python
# 상위 10개 데이터 출력하기
top_data = data[::-1][:10]
print(top_data)
```

```python
# 데이터 정보 확인(크기, 차원)
print(f'데이터 크기: {data.shape[0]}')
print(f'데이터 차원: {data.ndim}')
```

```python
# 전체 데이터 평균 평점 구하기
rating = np.mean(data[:, 2])
print(rating)
```

```python
# 사용자 아이디 1인 데이터 출력하기
user_1 = data[data[: , 0] == 1]
print(user_1)
```

```python
# 사용자 아이디 1인 사람의 평균 평점 구하기
rating_avg = np.mean(user_1[:,2])
print(rating_avg)
```

```python
# 각 사용자별 평균 평점 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
user_ids = np.unique(data[:, 0])
print(f'전체 사용자 수: {len(user_ids): ,}명')

user_avg_rating = [ [id, np.mean(data[data[:, 0] == id][:, 2])] for id in user_ids]
for id, avg_rating in user_avg_rating[:10]: # 10번 회원까지만 출력
    print(f'{id} : {avg_rating: .2f}명')
```

```python
np.set_printoptions(suppress=True, precision=2) # precision: 소수점 자리
```

```python
# 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기
user_avg_rating = np.array(user_avg_rating)
user_avg_rating[:10]
print(f'평균 평점이 4점 이상인 유저 수: {sum(user_avg_rating[:, 1] >= 4): ,}명')
```
