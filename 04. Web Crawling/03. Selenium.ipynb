## 구글에서 이미지 검색 후 검색 결과 6번 스크롤 하기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 페이지가 로드될때까지 기다리는 시간
SCROLL_PAUSE_TIME = 1
SEARCH_KEY_WORD = 'python'

# 웹드라이버 동적 다운로드 방식으로 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get('https://www.google.com/')

element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEY_WORD)
element.submit()

driver.find_element(By.LINK_TEXT, '이미지').click() # 링크로 검색할때는 생략 가능
time.sleep(SCROLL_PAUSE_TIME)
last_height = driver.execute_script("return document.body.scrollHeight")
print('last_height: ', last_height)

for i in range(6):
    # 윈도우 창을 0에서부터 가장 밑(scrollHeight)까지 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    # 더 이상 스크롤 될 페이지가 없을 경우 scrollHeight 값의 변화가 없다.
    print('new_height: ', new_height)
    print('-'*30)
```

## 구글에서 이미지 검색 후 검색 결과 무한 스크롤 - step(1)

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 페이지가 로드될때까지 기다리는 시간
SCROLL_PAUSE_TIME = 1
SEARCH_KEY_WORD = 'python'

# 웹드라이버 동적 다운로드 방식으로 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get('https://www.google.com/')

element = driver.find_element(by=By.CLASS_NAME, value='gLFyf') # 검색창
element.send_keys(SEARCH_KEY_WORD) # 'python' 검색
element.submit()

driver.find_element(By.LINK_TEXT, '이미지').click() # 링크로 검색할때는 생략 가능
time.sleep(SCROLL_PAUSE_TIME)
last_height = driver.execute_script("return document.body.scrollHeight")

# 스크롤 횟수
scroll_cnt = 1

while True:
    # 윈도우 창을 0에서부터 가장 밑(scrollHeight)까지 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    # 더 이상 스크롤 될 페이지가 없을 경우 scrollHeight 값의 변화가 없다.
    print(f'last Height: {last_height}, new Height: {new_height}, scroll count: {scroll_cnt}')

    if last_height != new_height: # 계속해서 new_height 값이 변경 된다면
        last_height = new_height
        scroll_cnt += 1
    else: # 더 이상 new_height 값이 변동이 없다면 ('결과 더보기'버튼 생성)
            element = driver.find_element(By.CSS_SELECTOR, '#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
            # 해당 엘리먼트와 상호작용이 불가한 상태가 되면 예외 발생
            try:
                element.click()
                print('결과 더보기 클릭!')
            except:
                break
```

## 구글에서 이미지 검색 후 검색 결과 무한 스크롤 - step(2)

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 페이지가 로드될때까지 기다리는 시간
SCROLL_PAUSE_TIME = 1
IMAGE_EXTRACTION_NUM = 10 # 이미지 추출 개수
SEARCH_KEY_WORD = 'python'

# 웹드라이버 동적 다운로드 방식으로 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get('https://www.google.com/')

element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEY_WORD)
element.submit()

driver.find_element(By.LINK_TEXT, '이미지').click() # 링크로 검색할때는 by, value 생략 가능
time.sleep(SCROLL_PAUSE_TIME)
last_height = driver.execute_script("return document.body.scrollHeight")

# 스크롤 횟수
scroll_cnt = 1

while True:
    # 윈도우 창을 0에서부터 가장 밑(scrollHeight)까지 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    # 더 이상 스크롤 될 페이지가 없을 경우 scrollHeight 값의 변화가 없다.
    print(f'last Height: {last_height}, new Height: {new_height}, scroll count: {scroll_cnt}')

    if last_height != new_height: # 계속해서 new_height 값이 변경 된다면
        last_height = new_height
        scroll_cnt += 1
    else: # 더 이상 new_height 값이 변동이 없다면 ('결과 더보기'버튼 생성)
            element = driver.find_element(By.CSS_SELECTOR, '#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
            # 해당 엘리먼트와 상호작용이 불가한 상태가 되면 예외 발생
            try:
                element.click()
                print('결과 더보기 클릭!')
            except:
                break

############### 이미지 선택 & 해당 이미지 src 저장 ###############
imgs = driver.find_elements(By.CSS_SELECTOR, '#islrg > div.islrc > div a.wXeWr.islib.nfEiy')

print('이미지 개수: ', len(imgs))
img_src_list = []
img_cnt = 0

for img in imgs:
    img.send_keys(Keys.ENTER)
    img_a = driver.find_element(By.CSS_SELECTOR, '#Sva75c > div.DyeYj > div > div.dFMRD > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div > div > div > div.n4hgof > div.MAtCL.b0vFpe > a')
    img_src = img_a.find_element(By.TAG_NAME, 'img').get_attribute('src')
    
    if not img_src.startswith('http'):
        continue
        
    img_cnt += 1
    print(f'[{img_cnt}].{img_src}')    
    img_src_list.append(img_src)
    
    if img_cnt == IMAGE_EXTRACTION_NUM: break

############### 검색 이미지 파일로 저장 ###############
import os
import requests

ts = time.localtime()
path = 'c:/Temp/'
now = f'{ts.tm_year}.{ts.tm_mon}.{ts.tm_mday}.{ts.tm_hour}.{ts.tm_min}.{ts.tm_sec}'
diretory = f'{SEARCH_KEY_WORD}/{now}/'
print(diretory)
os.chdir(path)
if not os.path.exists(diretory):
    os.makedirs(diretory)
    
file_no = 1
os.chdir(path+diretory)
for url in img_src_list:
    extension = url.split('.')[-1] # 원본 확장자
    ext = ''
    if extension in ['jpg', 'JPG', 'jpeg', 'JPEG', 'png', 'PNG', 'gif', 'GIF']:
        ext = '.'+extension
    else: # 확장자가 없을때
        ext = '.jpg'
        
    file_name = str(file_no)+'-'+SEARCH_KEY_WORD+ext
    file_no += 1
    res = requests.get(url)
        
    with open(file_name, 'wb') as f:
        f.write(res.content)
            
driver.close()
```
